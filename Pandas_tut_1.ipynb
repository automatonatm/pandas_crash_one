###  Apply Method

```python
df['len'] = df.Location.apply(len)
df
```

```python
def func(con):
    if con<10: return 'Good'
    elif con>10 & con<=50: return 'Poor'
    else: return 'Unsuitable'
    
def acidic(pH):
    if pH == 7: return 'Neutral'
    if pH > 7: return 'Basic'
    if pH < 7: return 'Acid'
    
```

```python
df['quality'] = df.AsIII.apply(func)
df['pH Range'] = df.pH.apply(acidic)
df
```

```python
df.groupby(['pH Range', 'pH'])['Location'].count()
```

```python
df['Finalized'] = df.AsIII.apply(lambda con: 'Good' if con > 10 else ('Poor' if con>10 & con<=50 else 'Unsuitable'))
df
```

```python
df['Witheld'] = df.Location.apply(lambda loc: True if (loc  == 'Aliabad') | (loc == 'Kaijuri') else False)
df
```

```python
df.groupby('Witheld')['Location'].count()
```
