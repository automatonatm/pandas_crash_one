#### conditional rendering

```python
df.loc[df['Ca']>88] #where ca concentration is > 88
```

```python
df.loc[df['Ca']>88, 'Ca']  #boolean indexing but fetch a row
```

```python
df.loc[(df['Ca']>88) & (df['SO4'] == 0.1)]  #multiple boolean
```

```python
pH = df.loc[(df['Ca']>88) & (df['SO4'] == 0.1), 'pH']  #multiple boolean use | for or
pH
```

```python
type(pH)
```

```python
df.loc[df['UNION'] == 'Aliabad', ['Mn','pH']] #Strings
```

```python
df.loc[df['UNION'].str.contains('Aliabad')] #df['UNION'].str.contains('Aliabad')
```
